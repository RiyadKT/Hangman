# Computing of $f(g,p,r)$

The following functions will be useful in the computing of f and throughout this notebook, we also provide randomly generated dictionaries for testing

```python

def tuple_to_dict(tup,distrib=None):
    "returns a dictionary of distribution distrib from tup"
    
    #uniform distribution in Case where no distribution is specified
    if distrib==None:
        distrib=len(tup)*[1/len(tup)]
        
    # Random distribution
    elif distrib=='random':
        tab=np.random.rand(len(tup))
        distrib=tab/np.sum(tab)
        distrib=list(distrib)
    res={}
    
    #Creating a dictionary with tuple entries as keys and distrib values as values
    distrib_copy=distrib.copy()
    for i in range(len(tup)):
        res[tup[i]]=distrib_copy.pop(0)
    return res  

tuple_to_dict((1,2),'random')
```

```python
#Same principle as tuple_to_dict (this function isn't even necessary)
def list_to_dict(tup,distrib=None):
    "returns a dictionary of distribution distrib from tup"
    if distrib==None:
        distrib=len(tup)*[1/len(tup)]
    elif distrib=='random':
        tab=np.random.rand(len(tup))
        distrib=tab/np.sum(tab)
        distrib=list(distrib)
    res={}
    distrib_copy=distrib.copy()
    for i in range(len(tup)):
        res[tup[i]]=distrib_copy.pop(0)
    return res  

list_to_dict([1,2],'random')
```

```python
def compute_letter_entropies(my_dict, my_alphabet):
    """
    Compute the entropy of the pattern associated to each letter, given a dictionary and an alphabet.
    """

    letter_entropies = []
    
    ### YOUR CODE HERE ###
    
    # Find words length from the dictionnary
    length = len(list(my_dict)[0])
    all_patterns = list(reversed(generate_all_patterns(length)))

    
    # For each letter
    for my_letter in my_alphabet:

        # Compute the patterns distribution
        pattern_probs = []
        words_seen=[]
        # For each pattern
        for my_pattern in all_patterns:
            p_pattern = 0

            # Filter the words that fit the pattern
            pattern_dict = filter_dict_by_pattern(my_letter, my_pattern, my_dict)
            if pattern_dict!={}:
                for word in pattern_dict.copy().keys():
                        if word in words_seen:
                            del pattern_dict[word]
                words_seen+=pattern_dict.keys()
            for item in pattern_dict.items():
                word, p_word = item
                p_pattern += p_word

            pattern_probs.append(p_pattern)


        # Normalise probabilities
        pattern_probs /= np.sum(pattern_probs)
        
        # Compute the entropy of that distribution
        pmf = np.array(pattern_probs)
        H_letter = -np.sum(pmf[pmf>0]*np.log2(pmf[pmf>0]))
        letter_entropies.append(H_letter)
        
    ### END YOUR CODE ###
        
    return letter_entropies

def list_entropies_to_dict(dicto,alphabet):
    "returns dictionary with each letter in alphabet and its corresponding entropy"
    l=compute_letter_entropies(dicto,alphabet)
    return dict([(alphabet[i],l[i]) for i in range(len(l))])
```

```python
          
import numpy as np

def winning_probability(g,p,r,pruning=True,distribution=None,return_memo=False):
    "returns f's value for g,p,r with words in p following distribution"
    #g=number of lives, p= set of words, r= set of letters
    #We do the following operations to take bigger masks into account first because if a word has an a at two spots, its corresponding a-mask will be the one at both spots and not the mask with only one a at either spot because the gamemaster has to reveal all spots containing a
    all_patterns = list(reversed(generate_all_patterns(5)))
    memo={}
    p_dict=list_to_dict(p,distribution)
    #Induction function
    def rec(i,j,k):
        #If we've already calculated the function at value i,j,k
        if (i,j,k) in list(memo.keys()):
            return memo[(i,j,k)]
        if i==0:
            memo[(i,j,k)]=0
            return 0
        if len(j)==1:
            memo[(i,j,k)]=1
            return 1
        else:
            #P
            dicto={k:v for (k,v) in p_dict.items() if k in j}
            #Finding word length
            if dicto!={}:
                length = len(list(dicto.keys())[0])
            else:
                length=0
            
            count=0
            
            #Entropy sorting
            list_entropies=list_entropies_to_dict(dicto,list(k))
            k_sorted=sorted(k,key=lambda x:list_entropies[x],reverse=True)
            
            m=0
            
            
            
            
            
            
            
            
            # Induction over letters
            for my_letter in k_sorted: 
                
                b=False
                
                
                words_seen=[]
                sum=0
                #used for pruning
                optimistic_cum_sum=np.zeros(len(all_patterns))
                cpt=0
                pattern_dict_list={}
                for pattern in all_patterns:
                    pattern_dict=filter_dict_by_pattern(my_letter,pattern,dicto)
                    pattern_dict={k:v for (k,v) in pattern_dict.items() if k not in words_seen}
                    words_seen+=pattern_dict.keys()
                    pattern_dict_list[tuple(pattern)]=pattern_dict
                    sum+= np.sum(list(pattern_dict.values())) 
                    if pruning:
                        
                        ##Very important to use np arrays or else computing time will explode
                        
                        optimistic_cum_sum[cpt]=sum
                        cpt+=1
                    
                
               
                # Calculating values for all masks
                mlist=[]

                
                words_seen=[]
                
                cpt=len(all_patterns)-1
                
                current_sum=0
                b=True

                for my_pattern in generate_all_patterns(5):
                    
                    #Pruning
                    if pruning and m!=0 and (current_sum + optimistic_cum_sum[cpt] <= m*sum ) : 
                        b=False
                        break
                    cpt-=1
                        
                    #Theta(my_pattern,P)
                    pattern_dict=pattern_dict_list[tuple(my_pattern)]
                    if pattern_dict!={}:
                    #Empty mask case
                        if my_pattern==[]:
                            c_val=1
                        else:
                            c_val=0
                        

                        #R->R\{r}
                
                        var=list(k)
                        var.remove(my_letter)
                        var=frozenset(var)

                        #recursive call
                        
                        
                        x=np.sum(list(pattern_dict.values()))*rec(i-c_val,tuple(pattern_dict.keys()),var)
                        current_sum+=x
                        if len(pattern_dict)!=0:
                            mlist.append(x)
                            count+=1
                            
                #If we stopped due to pruning then following insructions are useless
                if b :
                    mlist=np.array(mlist,dtype='float')
                    #normalising
                    sum_mlist=np.sum(mlist)/sum
                    m=max(m,sum_mlist)
                    if sum_mlist==1:
                        break
                

        
            memo[(i,j,k)]=m
            return m
    if return_memo:        
        return rec(g,tuple(p),frozenset(r)),memo
    else:
        return rec(g,tuple(p),frozenset(r))
                    
        
    
```

```python
          

def phi(p,r,pruning=True,distribution=None,return_memo=False,entropy_sorting=True,return_letters=False):
    "returns phi's value for p,r with words in p following distribution"
    #p= set of words, r= set of letters
    #We do the following operations to take bigger masks into account first because if a word has an a at two spots, its corresponding a-mask will be the one at both spots and not the mask with only one a at either spot because the gamemaster has to reveal all spots containing a
    letter_ouput={}
    all_patterns = list(reversed(generate_all_patterns(5)))
    memo={}
    p_dict=list_to_dict(p,distribution)
    #Induction function
    def rec(i,j):
        #If we've already calculated the function at value i,j,k
        if (i,j) in list(memo.keys()):
            return memo[(i,j)]
        if len(i)==1:
            memo[(i,j)]=1
            return 1
        else:
            #P
            dicto={k:v for (k,v) in p_dict.items() if k in i}
            
            
            #entropy sorting
            if entropy_sorting:
                list_entropies=list_entropies_to_dict(dicto,list(j))
                j_sorted=sorted(j,key=lambda x:list_entropies[x],reverse=True)
            
            
            else:
                j_sorted=j
                
                
                
            # For each letter
            m=np.inf
            
            
            
            
            
            
            
            argminletter=j_sorted[0]
            
            for my_letter in j_sorted: 
                
                b=False
                
                
                words_seen=[]
                sum=0
                #used for pruning
                optimistic_cum_sum=np.zeros(len(all_patterns))
                cpt=0
                pattern_dict_list={}
                for pattern in all_patterns:
                    pattern_dict=filter_dict_by_pattern(my_letter,pattern,dicto)
                    pattern_dict={k:v for (k,v) in pattern_dict.items() if k not in words_seen}
                    words_seen+=pattern_dict.keys()
                    pattern_dict_list[tuple(pattern)]=pattern_dict
                    sum+= np.sum(list(pattern_dict.values())) 
                    if pruning:
                        
                        ##Very important to use np arrays or else computing time will explode
                        
                        optimistic_cum_sum[cpt]=sum
                        cpt+=1
                    
               
                # Calculating values for all masks
                mlist=[]

                
                words_seen=[]
                
                cpt=len(all_patterns)-1
                
                current_sum=0
                b=True

                for my_pattern in generate_all_patterns(5):
                    
                    #Pruning
                    if pruning and m!=0 and (current_sum + optimistic_cum_sum[cpt] >= m*sum ) : 
                        b=False
                        break
                    cpt-=1
                        
                    #Theta(my_pattern,P)
                    pattern_dict=pattern_dict_list[tuple(my_pattern)]
                    if pattern_dict!={}:
                        

                        #R->R\{r}
                
                        var=list(j)
                        var.remove(my_letter)
                        var=frozenset(var)

                        #recursive call
                        
                        
                        x=np.sum(list(pattern_dict.values()))*rec(tuple(pattern_dict.keys()),var)
                        current_sum+=x
                        if len(pattern_dict)!=0:
                            mlist.append(x)
                
                if b :
                    mlist=np.array(mlist,dtype='float')
                    #normalising
                    sum_mlist=np.sum(mlist)/sum
                    mprev=m 
                    m=min(m,sum_mlist)
                    if m!=mprev:
                        argminletter=my_letter
            #This keeps track of what letter minimizes the sum
            letter_ouput[(i,j)]=argminletter
            memo[(i,j)]=m+1
            return m+1
    if return_memo:        
        return rec(tuple(p),frozenset(r)),memo
    elif return_letters:
        return rec(tuple(p),frozenset(r)),letter_ouput
    else:
        return rec(tuple(p),frozenset(r))
    
```

# Reinforcement learning agent

```python
class HangmanEnv:
    def __init__(self, word_list, weights,alphabet):
        self.weights=weights
        self.word_list = word_list
        self.alphabet = alphabet
        self.phitab=[]
        self.reset()
    
    def reset(self):
        self.word = random.choices(self.word_list, k=1)[0] #Choosing random word in dictionary
        self.turns_remaining = len( alphabet )
        self.possible_words = self.word_list.copy()
        self.possible_letters=alphabet.copy()
        return self.get_state()
    
    def get_state(self):
        return  self.possible_words,self.possible_letters
    
    def update_possible(self,letter,pattern):
        if self.possible_words!=[]:
           self.possible_words = list(filter_dict_by_pattern(letter,pattern,list_to_dict(self.possible_words)).keys())
        if letter in self.possible_letters:
            self.possible_letters.remove(letter)

    
    def step(self, letter):
        if len(self.possible_words)==1:
            return self.get_state(),1,True
        if letter not in self.possible_letters:
            return self.get_state(), -100, True  # Penalize for guessing an already guessed letter
        if letter in self.word:
            pattern=[]
            for idx, char in enumerate(self.word):
                if char == letter:
                    pattern.append(idx)
                    #self.guessed_word[idx] = letter
            #print(pattern)
            self.update_possible(letter,pattern)
            if len(self.possible_words)==1:
                return self.get_state(), 1, True  # Reward ending the game
            return self.get_state(), 3, False  # Reward for correct letter guess
        else:
            self.update_possible(letter,[])
            self.turns_remaining -= 1
            #self.update_possible_words()
            if self.turns_remaining == 0:
                return self.get_state(), -1, True  # Penalize for running out of turns
            return self.get_state(),  -1,False  # Penalize for incorrect letter guess

    #This is used in the evil hangman variant
    def step_evil(self,pattern):
        # transforming list to dictionnary without considering distribution as it does not matter here.
        letter=max(list_entropies_to_dict(list_to_dict(self.possible_words),self.possible_letters),key=list_entropies_to_dict(list_to_dict(self.possible_words),self.possible_letters).get)
        prev_words,prev_letters=self.possible_words,self.possible_letters
        self.update_possible(letter,pattern)
        if self.possible_words==[]:
            print(prev_words,prev_letters,letter,pattern)
        reward=1 # Case when player succesfully guesses letter
        if pattern==[]:
            self.turns_remaining-=1
            reward=2 # Player looses a life
        if self.turns_remaining==0:
            return self.get_state(),1,True #Made player loose
        if len(self.possible_words)==1:
            return self.get_state(),-1,True # Player guesses word
        return self.get_state(),reward,False
        
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

import torch
import torch.nn as nn

class HangmanLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=3):
        #Initialising 3 layer LSTM network
        super(HangmanLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, bidirectional=True)
        self.fc1 = nn.Linear(hidden_size * 2, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, output_size)
        self.relu = nn.ReLU()
        
    def forward(self, x, hidden):
        if len(x.shape) == 2:  # If input is unbatched (2D), add batch dimension
            x = x.unsqueeze(1)  # Make it (seq_len, 1, input_size)
        
        
        # Reshape hidden states if needed
        if hidden[0].size(1) != x.size(1):  # Check if hidden state batch size matches input batch size
            hidden = (hidden[0][:, :x.size(1), :], hidden[1][:, :x.size(1), :])
            
        out, hidden = self.lstm(x, hidden)
        out = out[:, -1, :]  # Take the output of the last time step
        out = self.fc1(out)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.relu(out)
        out = self.fc3(out)

        return out, hidden

    
    def init_hidden(self, batch_size, device):
        return (torch.zeros(self.num_layers * 2, batch_size, self.hidden_size).to(device),
                torch.zeros(self.num_layers * 2, batch_size, self.hidden_size).to(device))


```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque


class DQNAgent:
    def __init__(self, input_size, hidden_size, output_size, alphabet, word_list, batch_size=50, lr=1e-3, gamma=0.5, epsilon_start=1.0, epsilon_end=0.1, epsilon_decay=0.999, buffer_size=10000):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.alphabet = alphabet
        self.word_list = word_list
        self.output_size = output_size
        self.model = HangmanLSTM(input_size, hidden_size, output_size).to(self.device)
        self.target_model = HangmanLSTM(input_size, hidden_size, output_size).to(self.device)
        self.target_model.load_state_dict(self.model.state_dict())
        self.optimizer = optim.Adam(self.model.parameters(), lr=lr)
        self.gamma = gamma
        self.epsilon = epsilon_start
        self.epsilon_end = epsilon_end
        self.epsilon_decay = epsilon_decay
        self.replay_buffer = deque(maxlen=buffer_size)
        self.batch_size = batch_size
        self.update_target_every = 100
        self.steps = 0

    def select_action(self, state, possible_letters,hidden):
        state = torch.tensor(state, dtype=torch.float32).unsqueeze(0).to(self.device) 
        hidden = (hidden[0].to(self.device), hidden[1].to(self.device))

        if random.random() < self.epsilon: #Epsilon greedy policy
            action = random.choice([i for i in range(self.output_size) if chr(i + ord(self.alphabet[0])) in possible_letters])
        else:
            with torch.no_grad():
                #Passing through neural net to get q_values
                q_values, hidden = self.model(state, hidden)
                q_values = q_values.squeeze(0)  # Squeeze the batch dimension
            
            # Apply mask to zero out invalid actions
            mask = torch.tensor([1 if chr(i + ord(self.alphabet[0])) in possible_letters else 0 for i in range(self.output_size)], dtype=torch.float32).to(self.device)
            masked_q_values = q_values * mask
            
            # Make sure to handle the case where all actions are masked
            if mask.sum().item() == 0:
                action = random.choice([i for i in range(self.output_size)])
            else:
                action = masked_q_values.argmax().item()
                
        return action, hidden

    #This is used in the evil hangman variant
    def select_action_evil(self,state,possible_words,letter,hidden):
        #Same comments as select_action
        state = torch.tensor(state, dtype=torch.float32).unsqueeze(0).to(self.device)
        hidden = (hidden[0].to(self.device), hidden[1].to(self.device))
        b=0
        if random.random() < self.epsilon:
            action = random.choice([i for i in range(self.output_size) if (filter_dict_by_pattern(letter,patterns[i],list_to_dict(possible_words)) != {})])
            b=1
        else:
            with torch.no_grad():
                q_values, hidden = self.model(state, hidden)
                q_values = q_values.squeeze(0)  # Squeeze the batch dimension
            
            # Apply mask to zero out invalid actions
            mask = torch.tensor([1 if (filter_dict_by_pattern(letter,patterns[i],list_to_dict(possible_words)) != {}) else -np.inf for i in range(self.output_size)], dtype=torch.float32).to(self.device)
            masked_q_values = q_values * mask
            masked_q_values=torch.tensor([masked_q_values[i] if  abs(masked_q_values[i])!=np.inf else -np.inf  for i in range(len(masked_q_values))])
            
            
            # Make sure to handle the case where all actions are masked
            if mask.sum().item() == 0:
                action = random.choice([i for i in range(self.output_size) if (filter_dict_by_pattern(letter,patterns[i],list_to_dict(possible_words)) != {} )])
                b=2
                
            else:
                
                action = masked_q_values.argmax().item()
                b=3


        return action, hidden

    def store_experience(self, state, action, reward, next_state, done):
        self.replay_buffer.append((state, action, reward, next_state, done))

    def train_step(self, hidden):
        if len(self.replay_buffer) < self.batch_size:
            return

        batch = random.sample(self.replay_buffer, self.batch_size) #Taking sample from batch
        states, actions, rewards, next_states, dones = zip(*batch)

        
        states = torch.tensor(states, dtype=torch.float32).to(self.device)
        actions = torch.tensor(actions, dtype=torch.int64).to(self.device)
        rewards = torch.tensor(rewards, dtype=torch.float32).to(self.device)
        next_states = torch.tensor(next_states, dtype=torch.float32).to(self.device)
        dones = torch.tensor(dones, dtype=torch.float32).to(self.device)

        hidden = (hidden[0][:, :self.batch_size, :], hidden[1][:, :self.batch_size, :])  # Adjust hidden state batch size
        
        #Predicted q_values
        q_values, _ = self.model(states, hidden)
        q_values = q_values.gather(1, actions.unsqueeze(1)).squeeze(1)
        
        
        #Target q_values
        next_q_values, _ = self.target_model(next_states, hidden)
        next_q_values = next_q_values.max(1)[0]
        
    
        target_q_values = rewards + self.gamma * next_q_values * (1 - dones)


        
        loss = nn.MSELoss()(q_values, target_q_values) 
        
        #backprop/gradient descent
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()


        if self.epsilon > self.epsilon_end:
            self.epsilon *= self.epsilon_decay

        self.steps += 1
        if self.steps % self.update_target_every == 0:
            self.target_model.load_state_dict(self.model.state_dict())


    def encode_state(self, possible_words,possible_letters):
        #One hot encoding
        possible_letters_vector=[1 if letter in possible_letters else 0 for letter in self.alphabet]
        possible_words_vector = [1 if word in possible_words else 0 for word in self.word_list]
        return np.array(possible_words_vector + possible_letters_vector)
    
    
    #This is used in the evil hangman variant
    def encode_state_evil(self,possible_words,possible_letters):
        possible_words_vector = [1 if word in possible_words else 0 for word in self.word_list]
        possible_letters_vector=[1 if letter in possible_letters else 0 for letter in self.alphabet]
        letter=np.argmax(compute_letter_entropies(list_to_dict(possible_words),possible_letters))
        return np.array(possible_words_vector+possible_letters_vector+[1 if l==letter else 0 for l in self.alphabet])    
    
    



# Example usage
# Initialize environment, agent, and training loop
# Note: Define the `HangmanEnv` class according to your environment's logic

# Replace with actual word list
import json,random
with open('./normalised_freq_map.json', 'r') as f:
    full_dictionary = json.load(f)
word_list = random.choices(list(full_dictionary.keys()),k=int(len(full_dictionary)/100))
alphabet = [chr(97+i) for i in range(26)]
max_word_length = max(len(word) for word in word_list)
input_size = len(alphabet) + len(word_list)  # Max word length + size of the alphabet
hidden_size = int(2/3*input_size) + len (alphabet)
agent = DQNAgent(input_size, hidden_size, len(alphabet), alphabet, word_list)
word_weights=[full_dictionary[word] for word in word_list]
env = HangmanEnv(word_list,word_weights , alphabet)  # Replace with actual initialization
```
